```python
import numpy as np
```

## Creating toy dataset

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, random_state=53)
```

```python
X.shape
```

```python
plt.scatter(X.T[0], X.T[1], c=y)
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

## Decision boundaries for Logistic Regression

```python
model=LogisticRegression(solver='liblinear', C=0.1)
# model=DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_validate
```

```python
cross_validate(model, X, y, cv=10, return_train_score=True)['train_score'].mean()
```

```python
cross_validate(model, X, y, cv=10)['test_score'].mean()
```

```python
model.fit(X, y)
```

```python
x11, x22 = np.meshgrid(
    np.linspace(X.T[0].min(), X.T[0].max(), 1000),
    np.linspace(X.T[1].min(), X.T[1].max(), 1000)
)
```

```python
x11.shape
```

```python
x22.shape
```

```python
plt.scatter(X.T[0], X.T[1], c=y, edgecolors='black', linewidths=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.contourf(x11,
             x22,
             model.predict(
                 np.array(
                     [x11.ravel(),
                     x22.ravel()]
                         ).T).reshape(x11.shape), alpha=0.5)
```

Unless polynomial features are used, Logistic Regression model yields a linear decision boundary.

### Using `PolynomialFeatures` with Logistic Regression

At times, given features might come short to explain output, or the output might be a polynomial function of inputs.

If this is the case, `LogisticRegression` can be trained on polynomial features.

```python
model=LogisticRegression(solver='lbfgs', C=0.1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(5)
```

```python
poly.fit(X)
```

#### Cross validation to find the best polynomial degree

```python
for p in [2,3,4,5]:
    poly = PolynomialFeatures(p)
    poly.fit(X)
    score = cross_validate(model, poly.transform(X), y, cv=30)['test_score'].mean()
    ts = cross_validate(model, poly.transform(X), y, cv=30, return_train_score=True)['train_score'].mean()
    print('degree:', p, 'score:', score, 'train:', ts)
```

#### What `cross_validate` output looks like:

```python
cross_validate(model, poly.transform(X), y, cv=30)
```

```python
cross_validate(model, poly.transform(X), y, cv=30)['test_score']
```

### Fitting a Logistic Regression model using Polynomial Features

```python
model.fit(poly.fit_transform(X), y)
```

```python
x11, x22 = np.meshgrid(
    np.linspace(X.T[0].min(), X.T[0].max(), 1000),
    np.linspace(X.T[1].min(), X.T[1].max(), 1000)
)
```

```python
x11.shape
```

```python
x22.shape
```

```python
plt.scatter(X.T[0], X.T[1], c=y, edgecolors='black', linewidths=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.contourf(x11,
             x22,
             model.predict(
                 poly.transform(np.array(
                     [x11.ravel(),
                     x22.ravel()]
                         ).T)
             ).reshape(x11.shape), alpha=0.5)
```

## Probability Distribution

```python
plt.scatter(X.T[0], X.T[1], c=y, edgecolors='black', linewidths=2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.contourf(x11,
             x22,
             model.predict_proba(
                 np.array(
                     [x11.ravel(),
                     x22.ravel()]
                         ).T)[:,1].reshape(x11.shape), alpha=0.5, levels=np.linspace(0,1,100), cmap='jet', antialiasing=True)
plt.colorbar()
```
