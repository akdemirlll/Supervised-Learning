```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=40, n_features=2)
```

```python
plt.scatter(X.T[0], X.T[1], c=y)
plt.colorbar()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
model = KNeighborsRegressor(n_neighbors=1)
model.fit(X, y)
```

```python
x0range = np.linspace(X.T[0].min() - 1, X.T[0].max() + 1, 1000)
x1range = np.linspace(X.T[1].min() - 1, X.T[1].max() + 1, 1000)

xx, yy = np.meshgrid(x0range, x1range)

plt.figure(figsize=(8, 4))
plt.contourf(xx,
             yy,
             model.predict(np.array([xx.ravel(), yy.ravel()]).T).reshape(xx.shape),
             alpha=0.5,
             levels=np.linspace(-70, 70, 500)
            )
plt.scatter(X.T[0], X.T[1], c=y, edgecolors='black')
plt.colorbar()
_ = plt.title('K=1')
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_validate
```

```python
X, y = make_regression(n_samples=200, n_features=5)
```

```python
ks = list(range(1, 30))
trains = []
tests = []

for k in ks:
    knn = KNeighborsRegressor(n_neighbors=k)
    val = cross_validate(knn , X, y, cv=10, return_train_score=True)
    trains.append(val['train_score'].mean())
    tests.append(val['test_score'].mean())
```

```python
plt.figure(figsize=(14, 6))
plt.plot(ks, trains, label='train')
plt.plot(ks, tests, label='test')
plt.legend()
```

```python
np.random.seed(0)
z = np.random.randn(9, 2)
z

np.random.seed(0)
y = np.random.randint(2, size=9)

y

np.random.seed(5)
nokta = np.random.randn(1, 2)
nokta
```

```python
plt.scatter(z.T[0], z.T[1], c=y, label='z')
plt.scatter(nokta.T[0], nokta.T[1], color='red', label='nokta')
plt.legend()
```

```python
dist = np.sqrt(((z - nokta)**2).sum(axis=1))
```

```python
zz = z[dist.argsort()[:3]]
```

```python
plt.scatter(z.T[0], z.T[1], c=y, s=90, edgecolors='black')

plt.scatter(zz.T[0], zz.T[1], color='green')
plt.scatter(nokta.T[0], nokta.T[1], color='blue', label='nokta')
```

```python
y[dist.ravel().argsort()[:3]]
```

```python
def knn_classification(z, y, nokta, k):
    dist = np.sqrt(((z - nokta)**2).sum(axis=1))
    return (y[dist.argsort()[:k]].mean() > 0.5).astype(int)
```

```python
def knn_regression(z, y, nokta, k):
    dist = np.sqrt(((z - nokta)**2).sum(axis=1))
    return y[dist.argsort()[:k]].mean()
```

```python
knn_classification(z, y, nokta, 3)
```

```python
knn_classification(z, y, nokta, 5)
```

```python
knn_regression(z, y, nokta)
```
